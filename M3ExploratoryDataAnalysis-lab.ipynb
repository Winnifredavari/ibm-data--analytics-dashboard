```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
f = pd.read_csv("https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DA0321EN-SkillsNetwork/LargeData/m2_survey_data.csv")

# Create a box plot for the ConvertedComp column
plt.figure(figsize=(10, 6))
sns.boxplot(x=f['ConvertedComp'], color='skyblue')
plt.title('Box Plot of Converted Compensation')
plt.xlabel('Converted Compensation')
plt.grid(True)
plt.show()
```

```python
ax = sns.boxplot(x=df['Age'])
```

Find out the Inter Quartile Range for the column `ConvertedComp`.


```python
# your code goes here
IQR = df['ConvertedComp'].quantile(0.75) - df['ConvertedComp'].quantile(0.25)
IQR
```

Find out the upper and lower bounds.


```python
# your code goes here
#lower bounds
Lbound = df['ConvertedComp'].quantile(0.25) - 1.5*IQR
Lbound = max(Lbound, df['ConvertedComp'].min())
Lbound
```

```python
#upper bound 
Ubound = df['ConvertedComp'].quantile(0.75) + 1.5*IQR
Ubound
```

Identify how many outliers are there in the `ConvertedComp` column.


```python
# your code goes here
len(df[df['ConvertedComp'] > Ubound])
```

Create a new dataframe by removing the outliers from the `ConvertedComp` column.


```python
# your code goes here
df1 = df[df['ConvertedComp'] <= Ubound]
df1['ConvertedComp'].median()
```

```python
df1['ConvertedComp'].mean()
```
